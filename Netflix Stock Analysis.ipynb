```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime 
```

```python
df=pd.read_csv("Netflix.csv")
```

```python
sns.set(rc = {"figure.figsize" : (10,5)})
```

```python
df['Date'] = pd.to_datetime(df["Date"])
df=df.set_index("Date")
df.head()
```

```python
# Volume of stock trader
sns.lineplot(x=df.index,y=df.Volume,label="Volume")
plt.title("Volume of stock vs Time",fontweight = "bold")
plt.show()
```

```python
#stock price : High , Open , Close
df.plot(y=['High','Open','Close'])
plt.title('Netflix Stock Price',fontweight = 'bold',fontsize = 14)
plt.xlabel('Date',fontweight='bold',fontsize=10)
plt.ylabel('stock price',fontweight='bold',fontsize=10)
plt.show()
```

```python
#stock price : Day,Month,year wise

fig,(ax1,ax2,ax3)=plt.subplots(3,figsize=(15,10))

df.groupby(df.index.day).mean().plot(y='Volume',ax=ax1,xlabel='Day')
df.groupby(df.index.month).mean().plot(y='Volume',ax=ax2,xlabel='month')
df.groupby(df.index.year).mean().plot(y='Volume',ax=ax3,xlabel='Year')
plt.show()
```

```python
#Top 5 Dates with Highest Stock Price
a=df.sort_values(by='High',ascending=False).head(5)
a['High']
```

```python
# Top 5 dates with lowest price
b=df.sort_values(by='Low').head(5)
b['Low']
```

```python
#Trendline showing High and Low stock period of time

fig,axes=plt.subplots(nrows=1,ncols=2,sharex=True,figsize=(12,7))
fig.suptitle('High and Low Values Stock per period of time',fontweight='bold')
sns.lineplot(ax=axes[0],y=df['High'],x=df.index,color='green')
sns.lineplot(ax=axes[1],y=df['Low'],x=df.index,color='red')
```
