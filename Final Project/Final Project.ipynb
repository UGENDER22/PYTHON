```python
import json
class restaurant:
    def __init__(self):
        print("*******************WELCOME TO FOOD ZIZZY TOMATO APP*******************")
        self.foods={}
        self.user_foods={}
        self.food_id_=len(self.foods)+1
        
    def add_food_items(self):
        n=int(input("enter your items list:"))
        for i in range(1,n+1):
            self.name=input("Enter the food  ")
            self.quantity=int(input("Enter the quantity kg "))
            self.price=int(input("Enter the price "))
            self.discount=int(input("Enter Discount "))
            self.stock=int(input("Enter stock "))
            self.item={"name":self.name,"quantity":self.quantity,"price":self.price,"discount":self.discount,"stock":self.stock}
            self.food_id=len(self.foods)+1
            self.foods[self.food_id]=self.item
            self.user_items={"name":self.name,"quantity":self.quantity,"price":self.price}
            #self.user_food_id=len(self.user_items)+1
            self.user_foods[self.food_id]=self.user_items
            print(self.item)
            print(self.foods)
        with open("food_data.json","w") as f:
            json.dump(self.foods,f)
        with open("user_data.json","w") as d:
            json.dump(self.user_foods,d)
        print("items added successfully")
            
    def remove_food_items(self):
        n=int(input("enter the food id which you want to delete:"))
        del self.foods[n]
        print("Remaining items available",self.foods)
        
    def add_new_food_items(self):
        n=int(input("enter the new food item :"))
        for i in range(1,n+1):
            self.name=input("Enter the food  ")
            self.quantity=int(input("Enter the quantity kg "))
            self.price=int(input("Enter the price "))
            self.discount=int(input("Enter Discount "))
            self.stock=int(input("Enter stock "))
            self.item={"name":self.name,"quantity":self.quantity,"price":self.price,"discount":self.discount,"stock":self.stock}
            self.food_id=len(self.foods)+1
            self.foods[self.food_id]=self.item
            self.user_items={"name":self.name,"quantity":self.quantity,"price":self.price}
            self.user_food_id=len(self.user_items)+1
            self.user_foods[self.user_food_id]=self.user_items
            print(self.item)
            print(self.foods)
            with open("food_data.json","w+") as f:
                json.dump(self.foods,f)
            print("items added successfully")
            
    def edit_items_list(self):
        x = int(input("Enter food id that you want to update : \n"))
        if x in self.user_foods.keys():
            print("1. Update food name\n2. update quantity\n3. update price\n4. update discount\n5. update stock")
            y = input("Enter the key that you want edit : ")
            if y == "1":
                self.user_foods[x]["name"] = input("Enter your new food name : ")
                print("\n Successfully Updated \n")
            elif y == "2":
                self.user_foods[x]["quantity"] = float(input("Enter updated quantity : "))
                print("\n Successfully Updated \n")
            elif y == "3":
                self.user_foods[x]["price"] = int(input("Enter updated price : "))
                print("\n Successfully Updated \n")
            elif y == "4":
                self.user_foods[x]["discount"] = int(input("Enter the updated discount : "))
                print("\n Successfully Updated \n")
            elif y == "5":
                self.user_foods[x]["stock"] = int(input("Enter updated stock : "))
                print("\n Successfully Updated \n")
            else:
                print("Invalid option")
        else:
            print("Incorrect food id")
            
            
class user:
    def __init__(self):
        self.user_details={}
        self.update={}
        self.user_details_id=len(self.user_details)+1
        print("*************************************USER*********************************************")
    def user_registration(self):
        print("***********************WELCOME******************************")
        self.full_name=input("enter you name ")
        self.mobile_no=int(input("enter mobile no "))
        self.email_id=input("enter email_id ")
        self.address=input("enter your address ")
        self.password=input("enter your password ")
        user_details={"full_name":self.full_name,"mobile no":self.mobile_no,"email_id":self.email_id,"address":self.address,"password":self.password}
        self.user_details_id=len(user_details)+1
        self.user_details[self.user_details_id]=user_details
    def user_login(self):
        user_id=input("enter your user_id ")
        password=input("enter your password ")
        if user_id==self.email_id or  user_id==self.mobile_no:
            if password==self.password:
                print("SUCCESSFULLY LOGGED IN ")
                   
   
    def menu(self):
        print("1 . Place Your Order")
        print("2 . Order History")
        print("3 . Update Profile")
        with open ("user_data.json","r") as f:
            x = json.load(f)
        c = list(x.values())
        for i in c:
            print(i)
        n = int(input("Enter your option : "))
        if n <= 0:
            print("Enter valid option ")
        elif n == 1:
            print("Place your order")
            with open ("user_data.json","r") as d:
                x = json.load(d)
                print(x)
            n = [int(i) for i in input().split(",")]
            for j in x.keys():
                a = int(j)
                if a in n:
                    print(x[j])
                    u = x[j]
                    self.order_id = len(self.update)+1
                    self.update[self.order_id] = u
                    print("The items you ordered ", self.update)
            with open ("oder_history.json","w+") as f:
                json.dump(self.update,f)
        elif n == 2:
            print("order_history")
            with open ("oder_history.json","r") as f:
                a = json.load(f)
            d = list(a.values())
            for j in d:
                print(j)
        elif n == 3 :
            print("update profile")
            with open ("user_details.json","r") as f:
                x = json.load(f)
                print(x)
        elif n >= 4:
            print("Enter valid input")
            
    def update_profile(self):
        print("1. edit full_name\n2. edit mobile_no\n3. edit email_id\n4. edit address\n5. password\n6. exit")
        y=int(input("select your option "))
        if y==1:
            self.user_details["full_name"]=input("enter your new name ")
            print("DONE ")
        elif y==2:
            self.user_details["mobile_no"]=int(input("enter your new mobile no "))
            print("DONE")
        elif y==3:
            self.user_details["email_id"]=input("enter your new email_id ")
            print("DONE")
        elif y==4:
            self.user_details["address"]=input("enter your new address ")
            print("DONE ")
        elif y==5:
            self.user_details["password"]=input("enter your new password ")
            print("DONE ")
        else:
            print("EXIT")
       
            
        
        
        
              
            
                      
x=restaurant()
x.add_food_items()
x.remove_food_items()
x.edit_items_list()
y=user()
y.user_registration()
y.user_login()
y.menu()
y.update_profile()
```
