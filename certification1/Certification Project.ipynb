```python
import json
import re
class Amazon:
    def __init__(self):
        self.delivery_items = {}
        self.user_to_see = {}
        self.delivery_id = len(self.delivery_items)+1
        print("   -------------------------------------Welcome to our Amazon-------------------------------------------  ")
        print("                                                                                                             ")
    def items(self):
        number_of_items = int(input("Enter number of items to add : "))
        for i in range(1,number_of_items+1):
            self.Product_id = input("Enter Product id :")
            self.Product_name = input("Enter Product Name : ")
            self.manufacture=input("Enter Manufacture : ")
            self.price = int(input("Enter the price of item in Rupees : "))
            self.discount = int(input("Enter the discount in Rupees : "))
            self.stock = int(input("Enter how much stock left : "))
            self.list = {"Product_id":self.Product_id,"Product Name" : self.Product_name,"Manufacture":self.manufacture,"Price":self.price,"Discount":self.discount,"Stock":self.stock}
            self.delivery_id = len(self.delivery_items)+1
            self.delivery_items[self.delivery_id] = self.list
            self.user_list = {"Product id":self.Product_id,"Product Name" : self.Product_name,"Price":self.price}
            self.user_to_see[self.delivery_id] = self.user_list
        with open ("admin_list.json","w") as f:
            json.dump(self.delivery_items,f)
        with open ("user_list.json","w") as d:
            json.dump(self.user_to_see,d)
        print(self.delivery_items)
    def edit_items(self):
        x = int(input("Enter delivery id that you want to update : \n"))
        if x in self.delivery_items.keys():
            print("1.product id\n2. Product Name\n3. Update delivery name\n4.  update price\n5. update discount\n6. update stock")
            y = input("Enter the key that you want edit : ")
            if y == "1":
                self.delivery_items[x]["Product id"] = input("Enter your new Product id : ")
                print("\n Successfully Updated \n")
            elif y == "2":
                self.delivery_items[x]["Product Name"] = input("Enter updated Product Name : ")
                print("\n Successfully Updated \n")
            elif y == "3":
                self.delivery_items[x]["Delivery Name"] = int(input("Enter Delivery Name : "))
                print("\n Successfully Updated \n")
            elif y == "4":
                self.delivery_items[x]["Price"] = int(input("Enter updated price : "))
                print("\n Successfully Updated \n")
            elif y == "5":
                self.delivery_items[x]["Discount"] = int(input("Enter the updated discount : "))
                print("\n Successfully Updated \n")
            elif y == "6":
                self.delivery_items[x]["Stock"] = int(input("Enter updated stock : "))
                print("\n Successfully Updated \n")
            else:
                print("Invalid option")
        else:
            print("Incorrect delivery id")
    def add(self):
        number_of_items = int(input("Enter number of items to add : "))
        for i in range(1,number_of_items+1):
            self.Product_id = input("Enter product id : ")
            self.Product_name = input("Enter item Product Name : ")
            self.manufacture=input("Enter Manufacture name: ")
            self.price = int(input("Enter the price of item in Rupees : "))
            self.discount = int(input("Enter the discount in Rupees : "))
            self.stock = int(input("Enter how much stock left : "))
            self.list = {"Product id":self.Product_id,"Product Name" : self.Product_name,"manufacture":self.manufacture,"Price":self.price,"Discount":self.discount,"Stock":self.stock}
            self.delivery_id = len(self.delivery_items)+1
            self.delivery_items[self.delivery_id] = self.list
        with open ("admin_list.json","w+") as f:
            json.dump(self.delivery_items,f)
        print(self.delivery_items)
    def remove(self):
        item_id = int(input("Enter your item number to remove : "))
        del self.delivery_items[item_id]
        print(self.delivery_items)
class user(Exception):
    def __init__(self):
        self.user_details = {}
        self.personal = {}
        self.update = {}
        print("     -------------------------------------user--------------------------------------------------       ")
    def registration(self):
        self.name = input("Enter your name : ")
        mobile_no=input("Enter your Number : ")
        m=re.fullmatch("[6-9][0-9]{9}",mobile_no)
        if m!=None:
            print("Valid Mobile Number")
            self.mobile_no = mobile_no
            Email_id = input("Enter Your Mail Id : ")
            matcher = re.finditer("@",Email_id)
            y = 0
            for i in matcher:
                y=i.start()
            if Email_id[y+1:]=="gmail.com":
                print("It is a gmail Id")
                self.mail_id = Email_id
                self.address = input("Enter your Address : ")
                self.password = input("Enter your Password : ")
                self.user_details = {"Name":self.name,"Mobile no":self.mobile_no,"Mailid":self.mail_id,"Address":self.address,"Password":self.password}
                self.personal = {"Mobile no":self.mobile_no,"Password":self.password}
                print("user details updated successfully ")
        else :
            print("invalid number")
    def log_in(self):
        phone_number = int(input("Enter your number to login : "))
        password = input("Enter your Password to login : ")
        if phone_number == self.mobile_no:
            if password == self.password:
                print("Login successfull")
            else:
                print("invalid password")
    def menu(self):
        print("1 . Place Your Order")
        print("2 . Order History")
        print("3 . Update Profile")
        n = int(input("Enter your option : "))
        if n <= 0:
            print("Enter valid option ")
        elif n == 1:
            print("Place your order")
            with open ("user_list.json","r") as f:
                x = json.load(f)
            c = list(x.values())
            for i in c:
                print(i)
            with open ("user_list.json","r") as d:
                x = json.load(d)
            n = [int(i) for i in input().split(",")]
            for j in x.keys():
                a = int(j)
                if a in n:
                    print(x[j])
                    u = x[j]
                    self.order_id = len(self.update)+1
                    self.update[self.order_id] = u
                    print("The items you ordered ", self.update)
            with open ("oder_history.json","w+") as f:
                json.dump(self.update,f)
        elif n == 2:
            print("order_history")
            with open ("oder_history.json","r") as f:
                a = json.load(f)
            d = list(a.values())
            for j in d:
                print(j)
        else:
            print("Enter valid input")
    def update_profile(self):
        print("1. edit full_name\n2. edit mobile_no\n3. edit email_id\n4. edit address\n5. password\n6. exit")
        y=int(input("select your option "))
        if y==1:
            self.user_details["Name"]=input("enter your new name ")
            print("DONE ")
        elif y==2:
            self.user_details["Mobile_no"]=int(input("enter your new mobile no "))
            print("DONE")
        elif y==3:
            self.user_details["Mail_id"]=input("enter your new email_id ")
            print("DONE")
        elif y==4:
            self.user_details["Address"]=input("enter your new address ")
            print("DONE ")
        elif y==5:
            self.user_details["Password"]=input("enter your new password ")
            print("DONE ")
        elif y==6:
            return
        else:
            print("Enter valid number")
r = Amazon()
r.items()
r.add()
r.edit_items()
r.remove()
c = user()
c.registration()
c.log_in()
c.menu()
c.menu()
c.update_profile()
            
```
